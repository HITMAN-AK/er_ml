```python
import pandas as pd
```

```python
data=pd.read_csv("ecom.csv")
df=pd.DataFrame(data)
df[df.duplicated()]
```

```python
df.info()
```

```python
df[["name","des"]]=df["description"].str.split("-",n=1,expand=True)
df
```

```python
df.drop(columns='des',inplace=True)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['description'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)


```

```python
def get_top_similar_by_name(product_name, top_n=10):
    product_name = product_name.strip().lower()
    normalized_names = df['name'].str.strip().str.lower()
    if product_name not in normalized_names.values:
        return f"Product '{product_name}' not found in the dataset."
    product_index = normalized_names[normalized_names == product_name].index[0]
    similarity_scores = list(enumerate(cosine_sim[product_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    top_matches = similarity_scores[1:top_n + 1]
    return df.iloc[[match[0] for match in top_matches]]['name'].tolist()
input_product = "Active classic boxers"
top_similar_products = get_top_similar_by_name(input_product, top_n=10)
print(f"Top 10 products similar to '{input_product}':")
for s in top_similar_products:
    print(s)
```
